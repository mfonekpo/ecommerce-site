### Using Selenium to scrape the data

```python
browser = webdriver.Firefox(executable_path = firefox_driver_path) # using chrome webdriver
url = "https://webscraper.io/test-sites/e-commerce/static"

# open link with the browser
browser.get(url)

#save webpage content
page_content = browser.page_source

# parse the page content to BeautifulSoup
soup_doc = bs(page_content, "html.parser")

# navigating the webpage
browser.implicitly_wait(5) # sleep instance to makeup for the load time of some elements

computer_link = browser.find_element_by_xpath('//*[@id="side-menu"]/li[2]/a') #link to the computer tab
browser.execute_script("arguments[0].click();", computer_link)
time.sleep(5)
laptops_link = browser.find_element_by_xpath('//*[@id="side-menu"]/li[2]/ul/li[1]/a') #the laptop link button on the website
browser.execute_script("arguments[0].click();", laptops_link)
time.sleep(5)

#storing the data in a list
laptop_name = []
description = []
price = []
review = []
product_link = []
pages = 1

# A while loop to take care of the pagination. this helps automate the clicking of pages after scraping is done.
while (pages <= 20):

    print(f"Scraping data from page{pages}...")
    time.sleep(5)
    links = browser.find_elements_by_class_name("title")
    for link in links:
    #     getting product links
        product_link.append(link.get_property("href"))

    # Extracting the products data from the website
    thumbnail = browser.find_elements_by_class_name("thumbnail")
    for elems in thumbnail:
        products = elems.text.split("\n")
        laptop_name.append(products[1])
        price.append(products[0])
        description.append(products[2])
        review.append(products[3])
    print(f"done scraping data from page{pages}...\n\n")
    browser.implicitly_wait(5)
    next_page_link = browser.find_element_by_link_text("›")
    browser.execute_script("arguments[0].click();", next_page_link)
    pages += 1

# Breaking out of the loop once the page limit is reached...
else:
    print("\n\nDone scraping the website")
    print("closing browser now...")
    time.sleep(2)
    browser.close()
```

```python
product_catalogue = dict(
    laptop_name = laptop_name, 
    description = description,
    price = price,
    reviews = review,
    product_link = product_link
)
```

```python
products = pd.DataFrame(product_catalogue)
```

```python
products.to_csv("laptop_catalogue.csv", index = False)
```
